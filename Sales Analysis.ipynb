```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

## Data Cleaning

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

# Exploratory Data Analysis

## Gender Analysis

```python
ax = sns.countplot(x = 'Gender', data = df, hue='Gender')

for container in ax.containers:
    ax.bar_label(container)
```

```python

df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(x='Gender',y='Amount' ,data= sales_gen,hue='Gender')
for container in ax.containers:
    ax.bar_label(container)
```

##### From the above graphs we can see that most of the buyers are Females and even the purchasing power of Females is greater than men.

## Age Analysis

```python
df.columns
```

```python
ax = sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data= sales_age,hue='Gender')
```

##### From the above graphs we can see that most of the buyers are of age group between 26-35 yrs Females.

## State Analysis

```python
sales_state= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sns.barplot(data=sales_state,x='State' ,y='Orders',hue='State',dodge=False)
plt.xticks(rotation=45, ha='right')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(data=sales_state,x='State' ,y='Amount',hue='State',dodge=False)
plt.xticks(rotation=45, ha='right')
plt.show()
```

##### From the above graphs we can note that most of the orders are from Uttar Pradesh Maharashta,karnataka

## Marital Status

```python
df.columns
```

```python
ax = sns.countplot(data= df ,x= 'Marital_Status',hue = 'Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales_state,x='Marital_Status' ,y='Amount',hue='Gender')
```

##### From the above graphs we can see that most of the buyers are married (Women) and they have high purchasing power

## Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category')

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')  # ha='right' makes them slant nicely
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',hue='Product_Category')
```

##### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',hue='Product_ID')
```

## Conclusion:

#### Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.
